<h1>Python

<b>1.Matplotlib</b>: Matplotlib is one of the oldest and most widely used plotting libraries in Python. It provides a MATLAB-like interface and supports a wide variety of plots, including line plots, scatter plots, bar plots, histograms, and more. Matplotlib is highly customizable, allowing users to fine-tune every aspect of their visualizations.

```r
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

<b>2.Seaborn:</b> Seaborn is built on top of Matplotlib and provides a high-level interface for drawing attractive and informative statistical graphics. It simplifies the process of creating complex visualizations such as violin plots, box plots, and heatmaps. Seaborn also integrates well with Pandas data structures, making it a popular choice for exploratory data analysis.

```r
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot')
plt.show()
```

<b>3.Plotly:</b> Plotly is a versatile library that supports interactive, web-based visualizations. It can create a wide range of plots, including line charts, scatter plots, bar charts, and 3D plots. Plotly's interactive features allow users to zoom, pan, and hover over data points for additional information. Additionally, Plotly provides APIs for Python, R, and JavaScript, making it suitable for integration into web applications and dashboards.

```r
import plotly.graph_objects as go

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

fig = go.Figure(data=go.Scatter(x=x, y=y))
fig.update_layout(title='Line Plot', xaxis_title='X-axis', yaxis_title='Y-axis')
fig.show()
```

<h1>R

<b>1.ggplot2:</b> ggplot2 is one of the most widely used data visualization libraries in R. It is based on the grammar of graphics and provides a powerful and flexible system for creating a wide variety of plots. ggplot2 allows users to create plots layer by layer, making it easy to customize every aspect of the visualization. It supports a wide range of plot types, including scatter plots, bar plots, line plots, histograms, and more.

```r
library(ggplot2)


data <- data.frame(x = 1:10, y = rnorm(10))


ggplot(data, aes(x = x, y = y)) +
  geom_point() +
  xlab("X-axis") +
  ylab("Y-axis") +
  ggtitle("Scatter Plot")
```

<b>2.leaflet:</b> leaflet is a popular library for creating interactive maps in R. It provides a high-level interface for creating interactive maps with features such as markers, polygons, and popups. leaflet is based on the Leaflet JavaScript library and allows users to create custom maps with layers from various sources, including tile layers, GeoJSON, and shapefiles. leaflet is widely used for visualizing spatial data and creating interactive web-based maps.

```r
install.packages("leaflet")
library(leaflet)
```

```r
# load leaflet library
library(leaflet)

# create map widget with OpenStreetMap tile
map <- leaflet() %>%
	addTiles()

# print map widget
map
```

<b>3.dygraphs:</b> dygraphs is a library for creating interactive time series plots in R. It provides a high-level interface for creating interactive plots of time series data with features such as zooming, panning, and highlighting. dygraphs is based on the dygraphs JavaScript library and supports a wide range of customization options for visualizing time series data. It is commonly used for analyzing and visualizing time series data in fields such as finance, economics, and environmental science.

```r
install.packages("dygraphs")
```

```r
library(dygraphs)
data <- data.frame(Date = seq(as.Date("2022-01-01"), by = "day", length.out = 100), Value = rnorm(100))
dygraph(data, xlab = "Date", ylab = "Value", main = "Time Series Plot")
```
